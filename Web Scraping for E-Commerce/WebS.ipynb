```python
import requests
from bs4 import BeautifulSoup
import urllib3
import pandas as pd
import re
import numpy as np
```

```python
w_url= "https://www.flipkart.com/search?q=mobile"
http = urllib3.PoolManager()
response = http.request('GET', w_url)
data_web = BeautifulSoup(response.data.decode('utf-8'))
```

# Extracting Data of the First Page's top products

```python
f =open('flipcart_mobiles.csv',"wb")
f.write('name,star,ratings,reviews,cprice,mrp,storage,display,processor,imageUrl\n'.encode())
# Extracting all first page products
for bgbox in bg:
    product = bgbox.findAll('div',{'class':'_4rR01T'})
    name=product[0].text.replace(',','')
    
    stars = bgbox.find('div',{'class':"_3LWZlK"})
    try:
        star = stars.text
    except:
        star= 0 # for non star products

    r_ating  = bgbox.find('span',{'class':'_2_R_DZ'})
    try:
        rating_review = re.findall('\d+,*\d*',r_ating.text)
        ratings = rating_review[0].replace(',','')
        reviews = rating_review[1].replace(',','')
    except:
        ratings = 0
        reviews = 0
        
    # finding Current price
    cprice = bgbox.find('div', {'class':'_30jeq3 _1_WHN1'}).text.replace(',','').replace('₹','')
    
    
    
    
    # finding mrp price
    mrp = bgbox.find('div', {'class':'_3I9_wc _27UcVY'}).text.replace(',','').replace('₹','')
    try:
        mrp_price = mrp.text
    except:
        mrp_price = 0
    
    # finding informations
    detail = bgbox.findAll('li',{'class','rgWa7D'})
    storage = detail[0].text.replace(',','')
    display = detail[1].text.replace(',','')
    processor = detail[4].text.replace(',','')
    Image= bgbox.img
    imageUrl= Image.get('src')
    print (name,star,ratings,reviews,cprice,mrp,storage,display,processor,imageUrl)
    f.write(f"{name},{star},{ratings},{reviews},{cprice},{mrp},{storage},{display},{processor},{imageUrl}\n".encode())

    print('\n')
f.close()
```
